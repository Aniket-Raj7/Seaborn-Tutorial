```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset('tips')
df
```

```python
sns.scatterplot(x='tip',y ='total_bill',hue ="day", data = df,  palette = "YlGnBu" )
```

```python
sns.histplot(df['tip'], kde = True , bins = 30)
```

```python
sns.displot(df["tip"], kde = True,bins = 30)
```

```python
sns.barplot(x="sex", y ="tip", data = df, palette = 'YlGnBu')
```

```python
sns.boxplot(x='day', y = 'total_bill', data = df, hue = 'sex', palette = 'YlGnBu')
```

```python
sns.stripplot(x = "day" , y= "tip", data = df ,hue = "sex", palette = "YlGnBu")
```

```python
sns.jointplot(x = "tip", y = "total_bill",data=df, kind = 'hex', cmap= 'YlGnBu')
#kind = reg
```

```python
sns.jointplot(x = "tip", y = "total_bill",data=df, kind = 'kde', fill= True)
#kind = reg
```

```python
sns.jointplot(x = "tip", y = "total_bill",data=df, kind = 'reg')
#kind = reg
```
